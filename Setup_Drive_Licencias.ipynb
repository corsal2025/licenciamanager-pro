```python
import os

# Define la ruta raÃ­z en Google Drive
ROOT_PATH = '/content/drive/MyDrive/LICENCIA_MANAGER_SYSTEM'

# Define la estructura de carpetas
FOLDERS = [
    '01_ESCANEADAS_PENDIENTES/JUAN_PEREZ',
    '01_ESCANEADAS_PENDIENTES/MARIA_GONZALEZ',
    '01_ESCANEADAS_PENDIENTES/PEDRO_SOTO',
    '01_ESCANEADAS_PENDIENTES/ANA_ROJAS',
    '02_BASE_DATOS_EXCEL',
    '03_SUBIDAS_CONASET'
]

def create_structure():
    print(f"ðŸš€ Iniciando creaciÃ³n de estructura en: {ROOT_PATH}")

    # Crear RaÃ­z
    if not os.path.exists(ROOT_PATH):
        os.makedirs(ROOT_PATH)
        print(f"âœ¨ Carpeta RaÃ­z creada: {ROOT_PATH}")
    else:
        print(f"âœ… Carpeta RaÃ­z ya existe: {ROOT_PATH}")

    # Crear Subcarpetas
    for folder in FOLDERS:
        full_path = os.path.join(ROOT_PATH, folder)
        if not os.path.exists(full_path):
            os.makedirs(full_path)
            print(f"  âœ¨ Subcarpeta creada: {folder}")
        else:
            print(f"  âœ… Subcarpeta ya existe: {folder}")

    print("\nðŸŽ‰ Â¡Proceso completado exitosamente!")
    print("Ahora puedes ver la carpeta 'LICENCIA_MANAGER_SYSTEM' en tu Google Drive.")

create_structure()
```
